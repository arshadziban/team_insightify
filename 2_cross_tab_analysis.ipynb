## Physical Health Status

```python
df['C3_physical_problems'].value_counts()
```

| Category                  | Description                                                   |
|---------------------------|---------------------------------------------------------------|
| Headache Symptoms         | Records containing headache-related symptoms only             |
| Mood-related Symptoms     | Records containing irritability or mood-related symptoms only |
| Combined Symptoms         | Records containing both headache and irritability symptoms    |
| Other                     | Uncategorized or unspecified records                          |

```python
def categorize_symptoms(symptom):
    if pd.isna(symptom):
        return 'Other'
    symptom = str(symptom)
    # Check for combined symptoms first (contains both)
    if 'Headache' in symptom and 'Irritable mood' in symptom:
        return 'Combined Symptoms'
    # Then check for individual categories
    elif 'Headache' in symptom:
        return 'Headache Symptoms'
    elif 'Irritable mood' in symptom:
        return 'Mood-related Symptoms'
    else:
        return 'Other'

df['C3_physical_problems'] = df['C3_physical_problems'].apply(categorize_symptoms)
df['C3_physical_problems'].value_counts()
```

## Screen Content & AI Exposure 

```python
df['D4_content'].value_counts()
```

| Final Group                         | Included Categories |
|------------------------------------|---------------------|
| Group 1 – Educational Content      | Educational Content |
| Group 2 – Entertainment & Media    | Games, Games + Videos, Music, Videos, Videos + Social Media, Videos + Social Media + Educational Content |
| Group 3 – Social Media & Mixed Usage | Social Media, Games + Videos + Music + Social Media + Educational Content, Games + Videos + Music + Social Media + Video Chat + Educational Content + News |
| Other                              | Other |

```python
def categorize_content(content):
    if pd.isna(content):
        return 'Other'
    content = str(content).strip()
    
    # Group 1 – Educational Content
    if content == 'Educational Content':
        return 'Educational Content'
    
    # Group 2 – Entertainment & Media
    group2 = ['Games', 'Games + Videos', 'Music', 'Videos', 'Videos + Social Media', 
              'Videos + Social Media + Educational Content',
              # Handle comma-separated versions
              'Games, Videos', 'Music', 'Videos', 'Videos, Social Media',
              'Videos, Social Media, Educational Content']
    if content in group2:
        return 'Entertainment & Media'
    
    # Group 3 – Social Media & Mixed Usage
    group3 = ['Social Media', 'Games + Videos + Music + Social Media + Educational Content', 
              'Games + Videos + Music + Social Media + Video Chat + Educational Content + News',
              # Handle comma-separated versions
              'Games, Videos, Music, Social Media, Educational Content',
              'Games, Videos, Music, Social Media, Video Chat, Educational Content, News']
    if content in group3:
        return 'Social Media & Mixed Usage'
    
    # Other
    return 'Other'

df['D4_content_grouped'] = df['D4_content'].apply(categorize_content)
print("D4_content grouped:")
print(df['D4_content_grouped'].value_counts())
```
