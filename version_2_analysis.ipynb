## Physical Health Status

```python
df['C3_physical_problems'].value_counts()
```

| Category                  | Description                                                   |
|---------------------------|---------------------------------------------------------------|
| Headache Symptoms         | Records containing headache-related symptoms only             |
| Mood-related Symptoms     | Records containing irritability or mood-related symptoms only |
| Combined Symptoms         | Records containing both headache and irritability symptoms    |
| Other                     | Uncategorized or unspecified records                          |

```python
def categorize_symptoms(symptom):
    if pd.isna(symptom):
        return 'Other'
    symptom = str(symptom)
    # Check for combined symptoms first (contains both)
    if 'Headache' in symptom and 'Irritable mood' in symptom:
        return 'Combined Symptoms'
    # Then check for individual categories
    elif 'Headache' in symptom:
        return 'Headache Symptoms'
    elif 'Irritable mood' in symptom:
        return 'Mood-related Symptoms'
    else:
        return 'Other'

df['C3_physical_problems'] = df['C3_physical_problems'].apply(categorize_symptoms)
df['C3_physical_problems'].value_counts()
```
